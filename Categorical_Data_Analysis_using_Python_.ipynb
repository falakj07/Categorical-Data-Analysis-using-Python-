```python
import pandas as pd

data = {
    'Order_ID':['01','02','03','04','05','06','07','08','09','010'],

    'Category':['Electronics','Coding','Electronics','Grocery','Clothing',
                'Electronics','Grocery','Clothing','Electronics','Grocery'],

    'Payment Method': ["UPI","Card","UPI","Cash","Card","UPI","Cash","Card","UPI","Cash"],

    "Delivery_Type": ["Express","Standard","Express","Standard","Express","Standard","Express","Standard","Express","Standard"],

    "Customer_Type": ["New","Returning","Returning","New","Returning","New","Returning","New","Returning","New"],
    }

df=pd.DataFrame(data)
print(df)
```

```python
#1. Frequency Count
#Count how many orders in each category

print(df['Category'].value_counts())
print(df['Payment Method'].value_counts())
print(df['Order_ID'].value_counts())


```

```python
#2 . Unique Values
#Display all unique categories
print(df['Category'].unique())
```

```python
#3. Number of unique values
print(df['Category'].nunique())
```

```python
#4 .frequency of payment methods
print(df['Payment Method'].value_counts())
```

```python
#5 Cross tabulation
# category vs payment method

print(pd.crosstab(df['Category'],df['Payment Method']))
```

```python
#6. Cross Tabulation (Delivery vs Category Type)
print(pd.crosstab(df['Delivery_Type'],df['Category']))
```

```python
#7. percentage distrbution

print(df["Category"].value_counts(normalize=True)*100)
```

```python
#8. filtering Data
#Example : orders from electronics category
electronics_orders = df[df["Category"]=="Electronics"]
print(electronics_orders)
```

```python
#9. grouping
# Orders by grouping and payment method

print(df.groupby('Category')['Payment Method'].value_counts())
```

```python
## 10 sorting data
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```
